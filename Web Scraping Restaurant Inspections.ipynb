```python
def reportArray(link):
    database = requests.get(link).text
    soup = BeautifulSoup(database, 'html5lib')
    links = soup.find_all('a')
    for i in range(len(links)):
        links[i] = links[i].get('href')
    trueLinks = []
    for i in range(2, len(links)):
        trueLinks.append(links[i])
    reportsData = []
    for i in range(len(trueLinks)):
        string = trueLinks[i]
        upper = len(string)
        report = requests.get('http://il.healthinspections.us' + string[2:upper]).text
        soup = BeautifulSoup(report, 'html5lib')
        test = soup.find_all('td')
        check = True
        checkDate = False
        date = ''
        score1 = ''
        score2 = ''
        critical = ''
        repeats = ''
        for element in test:
            if checkDate:
                rawText = element.text
                rawText = rawText[len(rawText) - 8:len(rawText)]
                date = rawText
                checkDate = False
            if element.find(text=re.compile('SCORE')):
                if check:
                    score1 = element.text
                    check = not check
                else:
                    score2 = element.text
            if element.find(text=re.compile('Date')):
                checkDate = True
            if element.find(text=re.compile('CRITICAL X')):
                critical = element.text
            if element.find(text=re.compile('REPEATS X')):
                repeats = element.text
        reportsData.append([date, score1, score2, critical, repeats])
    return reportsData

print(reportArray('http://il.healthinspections.us/champaign/estab.cfm?facilityID=781'))
```
